```python
import pandas as pd
import numpy as np
from scipy import stats
df=pd.read_csv("train-selected-columns.csv")
print("-------DATASET INFORMATION-------")
target_column = "Age"
data=df[target_column].dropna()
print(" mean values is:",np.mean(data))
print(" median value is:",np.median(data))
print(" mode value is:",stats.mode(data))
print("variance value is:",np.var(data))
print("standard deviation value is:",np.std(data))
q1=np.percentile(data,25)  
print(" Q1(%) is:",q1)
q2=np.percentile(data,50)
print(" Q2(%) is:",q2)
q3=np.percentile(data,75)
print(" Q3(%) is:",q3)
inr=q3-q1
print("interquartile range is :",inr)
```
